```python
import numpy as np
import random as random
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
class Randwalk2D():

    ''' 
     This is a class to create on Two Dimentional Walk:
     x0,y0 : initial positions 
     N :  number of steps in random walk 
    '''

    def __init__(self,N,pos):
        self.N = N
        self.a0 = pos[0]
        self.b0 = pos[1]       
        self.X = []
        self.Y = []
        
        
    def move2D(self,x0,y0):
        '''
        This function choose a direction and walk one step in 2D
        '''
        direction =  random.choice(['x','y'])
        if direction == 'x':
            r = random.choice([-1,1])
            if r ==1:xf = x0+1
            else: xf = x0-1
            yf = y0
        if direction == 'y':
            r = random.choice([-1,1])
            if r ==1:yf = y0+1
            else: yf = y0-1
            xf = x0
        return xf,yf
        

    def walk2D(self):
        '''note initial position and variables to assign 2D walk informations'''
        x0 = self.a0
        y0 = self.b0
        X  = self.X
        Y  = self.Y
        
        k=0
        while k < self.N:
            '''move a step'''
            x1,y1 = self.move2D(x0,y0)
            #print(x1)
            '''collect X coordinate'''
            X.append(x1)
        
            '''collect Y coordinate'''
            Y.append(y1)
        
            '''set previous position for next step'''
            x0 = x1
            y0 = y1
        
            k = k+1
        
        return X,Y
```

```python
N = 100000
U = Randwalk2D(N,[1,1]).walk2D()
V = Randwalk2D(N,[1,1]).walk2D()
W = Randwalk2D(N,[1,1]).walk2D()

plt.figure(figsize = [5,5])
plt.plot(U[0],U[1],".",color='pink')
plt.plot(V[0],V[1],".",color = 'orange')
plt.plot(W[0],W[1],".", color = 'green')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.title("Brownian motion in 2D")
plt.show()
```
